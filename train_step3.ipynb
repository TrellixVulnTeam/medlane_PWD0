```python
import os
from Seq2seq import build
os.environ['CUDA_LAUNCH_BLOCKING'] = '1'
seq2seq, optimizer, Tensor, train_dataloader, val_set, test_set, loss_fun, config = build(768, 16, 64, True)
```

```python
from Seq2seq import eval_set
import torch
```

```python
tokens = config.tokenizer.tokenize('[CLS] config [MASK] [SEP]')
print(tokens)
ids = config.tokenizer.convert_tokens_to_ids(tokens)
print(ids)
print(config.tokenizer.convert_ids_to_tokens(range(10)))
```

```python
seq2seq.eval()
results = []
references = []
dics = []
for i, (batch_src, train_tar, train_dic) in enumerate(val_set):
    decoder_outputs, decoder_hidden, ret_dict = seq2seq(batch_src, None, 0.0)
    symbols = ret_dict['sequence']
    symbols = torch.cat(symbols, 1).data.cpu().numpy()
    sentences = decode_sentence(symbols, config)
    results += sentences
    references += train_tar
    dics += train_dic
```

```python
save_file_best = torch.load('./cache/best_save.data')
seq2seq.load_state_dict(save_file_best['para'])
test_results, bleu, hit, com = eval_set(seq2seq, val_set, config)
print('Test BLEU:%f, HIT:%f, COMMON:%f' % (bleu, hit, com))
```

```python
with open('./result/test_results.txt','w', encoding = 'utf-8') as f:
    for u,v,w in zip(test_results, src, tar):
        temp = v.split('[ unused2 ]')
        x = ''
        for t in temp:
            x = x+t.strip()+' '
        x = x.strip()
        f.write('step2: '+x[6:]+'\n')
        f.write('src  : '+v[7:]+'\n')
        f.write('tar  : '+w[7:]+'\n')
        f.write('step3: '+u+'\n')
        f.write('\n')
```
